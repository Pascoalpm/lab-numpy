```python
# Упражнение 1 - Создание и изменение массивов
import numpy as np

# 1. Создаем массив 3x3 с числами от 1 до 9
originalniy_array = np.arange(1, 10).reshape(3, 3)
print("=== ИСХОДНЫЙ МАССИВ ===")
print(originalniy_array)
print(f"Размерность: {originalniy_array.shape}")
print()

# 2. Умножаем все элементы на 2
umnozhenniy_array = originalniy_array * 2
print("=== ПОСЛЕ УМНОЖЕНИЯ НА 2 ===")
print(umnozhenniy_array)
print()

# 3. Заменяем элементы больше 10 на 0
konechniy_array = umnozhenniy_array.copy()  # Создаем копию
konechniy_array[konechniy_array > 10] = 0
print("=== ИТОГОВЫЙ МАССИВ (элементы > 10 заменены на 0) ===")
print(konechniy_array)
print()

# 4. Дополнительные проверки
print("=== ПРОВЕРКИ ===")
print(f"Исходные элементы: {originalniy_array.flatten()}")
print(f"После умножения: {umnozhenniy_array.flatten()}")
print(f"Итоговые элементы: {konechniy_array.flatten()}")
print(f"Количество нулей в итоговом массиве: {np.sum(konechniy_array == 0)}")
```
