```python
# Упражнение 8 - Определитель и обратная матрица
import numpy as np

# 1. Создаем случайную квадратную матрицу 3×3
np.random.seed(42)  # Для воспроизводимости результатов
matrica = np.random.randint(1, 11, size=(3, 3))

print("=== ИСХОДНАЯ МАТРИЦА 3×3 ===")
print(matrica)
print(f"Размерность: {matrica.shape}")
print()

# 2. Вычисляем определитель
determinant = np.linalg.det(matrica)
print("=== ОПРЕДЕЛИТЕЛЬ МАТРИЦЫ ===")
print(f"det(A) = {determinant:.2f}")
print()

# 3. Проверяем, существует ли обратная матрица
# Обратная матрица существует только если определитель ≠ 0
print("=== ПРОВЕРКА СУЩЕСТВОВАНИЯ ОБРАТНОЙ МАТРИЦЫ ===")
if abs(determinant) < 1e-10:  # Проверка с учетом погрешности вычислений
    print("Определитель близок к нулю. Матрица вырождена.")
    print("Обратная матрица НЕ существует.")
else:
    print(f"|det(A)| = {abs(determinant):.2f} ≠ 0")
    print("Обратная матрица СУЩЕСТВУЕТ.")
    print()
    
    # 4. Вычисляем обратную матрицу
    obratnaya = np.linalg.inv(matrica)
    print("=== ОБРАТНАЯ МАТРИЦА A^(-1) ===")
    print(obratnaya)
    print()
    
    # 5. Проверка: A × A^(-1) должна равняться единичной матрице
    print("=== ПРОВЕРКА: A × A^(-1) ===")
    proizvedenie = matrica @ obratnaya
    print(proizvedenie)
    print()
    
    # Округляем для проверки (из-за погрешностей вычислений)
    proizvedenie_okrugl = np.round(proizvedenie, 5)
    print("Проверка (с округлением до 5 знаков):")
    print(proizvedenie_okrugl)
    
    # Проверяем, близка ли к единичной матрице
    edinichnaya = np.eye(3)
    print(f"\nA × A^(-1) ≈ E ? {np.allclose(proizvedenie, edinichnaya)}")
    
    # 6. Проверка: A^(-1) × A тоже должна равняться единичной матрице
    print(f"\nA^(-1) × A ≈ E ? {np.allclose(obratnaya @ matrica, edinichnaya)}")
print()

# 7. Дополнительная информация
print("=== ДОПОЛНИТЕЛЬНАЯ ИНФОРМАЦИЯ ===")
print("Свойства определителя:")
print(f"det(A) = {determinant:.2f}")
print(f"det(A^(-1)) = {np.linalg.det(obratnaya):.2f}")
print(f"det(A) × det(A^(-1)) = {determinant * np.linalg.det(obratnaya):.2f} (должно быть 1)")
```
