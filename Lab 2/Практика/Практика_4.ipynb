```python
# Упражнение 4 - Генерация и работа с линейными последовательностями
import numpy as np

# 1. Создаем массив из 50 чисел, равномерно распределенных от -10 до 10
# Используем np.linspace() для равномерного распределения
array = np.linspace(-10, 10, 50)

print("=== ИСХОДНЫЙ МАССИВ ===")
print(f"Первые 10 элементов: {array[:10]}")
print(f"Последние 10 элементов: {array[-10:]}")
print(f"Размерность массива: {array.shape}")
print(f"Количество элементов: {len(array)}")
print(f"Минимальное значение: {np.min(array):.2f}")
print(f"Максимальное значение: {np.max(array):.2f}")
print()

# 2. Вычисляем сумму всех элементов
summa_vseh = np.sum(array)
print("=== СУММА ВСЕХ ЭЛЕМЕНТОВ ===")
print(f"Сумма всех элементов: {summa_vseh:.2f}")
print()

# 3. Вычисляем сумму положительных элементов (> 0)
polozhitelnye = array[array > 0]
summa_polozhitelnyh = np.sum(polozhitelnye)
print("=== СУММА ПОЛОЖИТЕЛЬНЫХ ЭЛЕМЕНТОВ ===")
print(f"Положительные элементы (первые 10): {polozhitelnye[:10]}")
print(f"Количество положительных элементов: {len(polozhitelnye)}")
print(f"Сумма положительных элементов: {summa_polozhitelnyh:.2f}")
print()

# 4. Вычисляем сумму отрицательных элементов (< 0)
otricatelnye = array[array < 0]
summa_otricatelnyh = np.sum(otricatelnye)
print("=== СУММА ОТРИЦАТЕЛЬНЫХ ЭЛЕМЕНТОВ ===")
print(f"Отрицательные элементы (первые 10): {otricatelnye[:10]}")
print(f"Количество отрицательных элементов: {len(otricatelnye)}")
print(f"Сумма отрицательных элементов: {summa_otricatelnyh:.2f}")
print()

# 5. Проверка: сумма всех = сумма положительных + сумма отрицательных
print("=== ПРОВЕРКА ===")
print(f"Сумма всех элементов: {summa_vseh:.2f}")
print(f"Сумма положительных + сумма отрицательных: {summa_polozhitelnyh + summa_otricatelnyh:.2f}")
print(f"Равенство выполняется: {np.isclose(summa_vseh, summa_polozhitelnyh + summa_otricatelnyh)}")
print()

# 6. Дополнительная статистика
print("=== ДОПОЛНИТЕЛЬНАЯ СТАТИСТИКА ===")
print(f"Среднее значение: {np.mean(array):.2f}")
print(f"Медиана: {np.median(array):.2f}")
print(f"Стандартное отклонение: {np.std(array):.2f}")
print(f"Дисперсия: {np.var(array):.2f}")
```
