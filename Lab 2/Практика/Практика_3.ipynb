```python
# Упражнение 3 - Объединение и разделение массивов
import numpy as np

# 1. Создаем два массива 1x5 со случайными числами от 0 до 50
np.random.seed(42)  # Для воспроизводимости результатов
array1 = np.random.randint(0, 51, size=5)
array2 = np.random.randint(0, 51, size=5)

print("=== ИСХОДНЫЕ МАССИВЫ ===")
print(f"Массив 1: {array1}")
print(f"Массив 2: {array2}")
print(f"Размерность массива 1: {array1.shape}")
print(f"Размерность массива 2: {array2.shape}")
print()

# 2. Объединяем массивы в один двумерный (по строкам)
# Используем np.vstack() для вертикального объединения (по строкам)
obedinenny = np.vstack((array1, array2))

print("=== ОБЪЕДИНЕННЫЙ МАССИВ (по строкам) ===")
print(obedinenny)
print(f"Размерность объединенного массива: {obedinenny.shape}")
print(f"Количество строк: {obedinenny.shape[0]}")
print(f"Количество столбцов: {obedinenny.shape[1]}")
print()

# 3. Разделяем полученный массив на два массива по 5 элементов
# Используем np.vsplit() для разделения по вертикали
array1_razdelenny, array2_razdelenny = np.vsplit(obedinenny, 2)

print("=== РАЗДЕЛЕННЫЕ МАССИВЫ ===")
print("Первый разделенный массив:")
print(array1_razdelenny)
print(f"Размерность: {array1_razdelenny.shape}")

print("\nВторой разделенный массив:")
print(array2_razdelenny)
print(f"Размерность: {array2_razdelenny.shape}")
print()

# 4. Преобразуем обратно в одномерные для сравнения с оригиналом
print("=== СРАВНЕНИЕ С ОРИГИНАЛОМ ===")
print(f"Оригинальный массив 1: {array1}")
print(f"Разделенный массив 1 (преобразованный): {array1_razdelenny.flatten()}")
print(f"Оригинальный массив 2: {array2}")
print(f"Разделенный массив 2 (преобразованный): {array2_razdelenny.flatten()}")
print()

# 5. Проверка: равны ли оригиналы и разделенные массивы
print("=== ПРОВЕРКА СООТВЕТСТВИЯ ===")
print(f"Массив 1 совпадает с оригиналом: {np.array_equal(array1, array1_razdelenny.flatten())}")
print(f"Массив 2 совпадает с оригиналом: {np.array_equal(array2, array2_razdelenny.flatten())}")
```
