```python
# Упражнение 5 - Работа с диагональными и единичными матрицами
import numpy as np

# 1. Создаем единичную матрицу 4x4
edinichnaya = np.eye(4)
print("=== ЕДИНИЧНАЯ МАТРИЦА 4x4 ===")
print(edinichnaya)
print(f"Размерность: {edinichnaya.shape}")
print()

# 2. Создаем диагональную матрицу 4x4 с элементами [5, 10, 15, 20]
# Используем np.diag() для создания диагональной матрицы
diagonalnaya = np.diag([5, 10, 15, 20])
print("=== ДИАГОНАЛЬНАЯ МАТРИЦА 4x4 ===")
print(diagonalnaya)
print(f"Размерность: {diagonalnaya.shape}")
print()

# 3. Находим сумму всех элементов единичной матрицы
summa_edinichnaya = np.sum(edinichnaya)
print("=== СУММА ЭЛЕМЕНТОВ ЕДИНИЧНОЙ МАТРИЦЫ ===")
print(f"Сумма всех элементов: {summa_edinichnaya}")
print(f"Сумма элементов на главной диагонали (след): {np.trace(edinichnaya)}")
print()

# 4. Находим сумму всех элементов диагональной матрицы
summa_diagonalnaya = np.sum(diagonalnaya)
print("=== СУММА ЭЛЕМЕНТОВ ДИАГОНАЛЬНОЙ МАТРИЦЫ ===")
print(f"Сумма всех элементов: {summa_diagonalnaya}")
print(f"Сумма элементов на главной диагонали (след): {np.trace(diagonalnaya)}")
print()

# 5. Сравниваем результаты
print("=== СРАВНЕНИЕ РЕЗУЛЬТАТОВ ===")
print(f"Сумма элементов единичной матрицы: {summa_edinichnaya}")
print(f"Сумма элементов диагональной матрицы: {summa_diagonalnaya}")
print(f"Разница: {abs(summa_diagonalnaya - summa_edinichnaya)}")
print()

# 6. Детальный анализ
print("=== ДЕТАЛЬНЫЙ АНАЛИЗ ===")
print("Единичная матрица - элементы на диагонали:")
print(np.diag(edinichnaya))
print("Единичная матрица - элементы вне диагонали:")
print(f"Количество нулей вне диагонали: {np.sum(edinichnaya == 0)}")

print("\nДиагональная матрица - элементы на диагонали:")
print(np.diag(diagonalnaya))
print("Диагональная матрица - элементы вне диагонали:")
print(f"Количество нулей вне диагонали: {np.sum(diagonalnaya == 0)}")
print()

# 7. Альтернативные способы создания
print("=== АЛЬТЕРНАТИВНЫЕ СПОСОБЫ СОЗДАНИЯ ===")

# Единичная матрица через identity
edinichnaya_alt = np.identity(4)
print("Единичная через np.identity(4):")
print(edinichnaya_alt)

# Диагональная матрица через zeros и заполнение диагонали
diagonalnaya_alt = np.zeros((4, 4))
np.fill_diagonal(diagonalnaya_alt, [5, 10, 15, 20])
print("\nДиагональная через np.zeros и fill_diagonal:")
print(diagonalnaya_alt)
```
