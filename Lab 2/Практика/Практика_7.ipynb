```python
# Упражнение 7 - Умножение матриц
import numpy as np

# 1. Создаем матрицы
np.random.seed(42)  # Для воспроизводимости результатов

# Первая матрица: 2 строки, 3 столбца (2×3)
matrica_A = np.random.randint(1, 11, size=(2, 3))
print("=== МАТРИЦА A (2×3) ===")
print(matrica_A)
print(f"Размерность: {matrica_A.shape}")
print(f"Строк: {matrica_A.shape[0]}, Столбцов: {matrica_A.shape[1]}")
print()

# Вторая матрица: 3 строки, 2 столбца (3×2)
matrica_B = np.random.randint(1, 11, size=(3, 2))
print("=== МАТРИЦА B (3×2) ===")
print(matrica_B)
print(f"Размерность: {matrica_B.shape}")
print(f"Строк: {matrica_B.shape[0]}, Столбцов: {matrica_B.shape[1]}")
print()

# Проверка условия умножения
print("=== УСЛОВИЕ УМНОЖЕНИЯ ===")
print(f"Столбцов в A: {matrica_A.shape[1]}")
print(f"Строк в B: {matrica_B.shape[0]}")
print(f"Можно умножать? {matrica_A.shape[1] == matrica_B.shape[0]}")
print()

# 2. Матричное умножение (разные способы)
print("=== МАТРИЧНОЕ УМНОЖЕНИЕ ===")

# Способ 1: оператор @ (рекомендуемый)
rezultat_1 = matrica_A @ matrica_B
print("Способ 1 (A @ B):")
print(rezultat_1)
print(f"Размерность результата: {rezultat_1.shape}")  # Должно быть 2×2
print()

# Способ 2: функция np.dot()
rezultat_2 = np.dot(matrica_A, matrica_B)
print("Способ 2 (np.dot(A, B)):")
print(rezultat_2)
print()

# Способ 3: метод .dot()
rezultat_3 = matrica_A.dot(matrica_B)
print("Способ 3 (A.dot(B)):")
print(rezultat_3)
print()

# 3. Детальный расчет для понимания
print("=== ДЕТАЛЬНЫЙ РАСЧЕТ ===")
print("Результирующая матрица C (2×2):")
print("C[0,0] = A[0,0]×B[0,0] + A[0,1]×B[1,0] + A[0,2]×B[2,0]")
print(f"       = {matrica_A[0,0]}×{matrica_B[0,0]} + {matrica_A[0,1]}×{matrica_B[1,0]} + {matrica_A[0,2]}×{matrica_B[2,0]}")
print(f"       = {matrica_A[0,0] * matrica_B[0,0]} + {matrica_A[0,1] * matrica_B[1,0]} + {matrica_A[0,2] * matrica_B[2,0]}")
print(f"       = {rezultat_1[0,0]}")
print()

print("C[0,1] = A[0,0]×B[0,1] + A[0,1]×B[1,1] + A[0,2]×B[2,1]")
print(f"       = {matrica_A[0,0]}×{matrica_B[0,1]} + {matrica_A[0,1]}×{matrica_B[1,1]} + {matrica_A[0,2]}×{matrica_B[2,1]}")
print(f"       = {matrica_A[0,0] * matrica_B[0,1]} + {matrica_A[0,1] * matrica_B[1,1]} + {matrica_A[0,2] * matrica_B[2,1]}")
print(f"       = {rezultat_1[0,1]}")
print()

# 4. Сравнение с поэлементным умножением
print("=== СРАВНЕНИЕ С ПОЭЛЕМЕНТНЫМ УМНОЖЕНИЕМ ===")
print("Поэлементное умножение НЕВОЗМОЖНО, так как размерности не совпадают!")
print("Для поэлементного умножения нужны матрицы одинакового размера.")
print()

# 5. Дополнительные примеры
print("=== ДОПОЛНИТЕЛЬНЫЕ ПРИМЕРЫ ===")

# Транспонирование и умножение
print("A × B =")
print(rezultat_1)
print("\nB^T × A^T =")
print(matrica_B.T @ matrica_A.T)
print("(B^T × A^T) == (A × B)^T?", np.array_equal(matrica_B.T @ matrica_A.T, (rezultat_1).T))
```
