```python
# Упражнение 9 - Транспонирование и след матрицы
import numpy as np

# 1. Создаем матрицу 4×4 со случайными целыми числами от 1 до 50
np.random.seed(42)  # Для воспроизводимости результатов
matrica = np.random.randint(1, 51, size=(4, 4))

print("=== ИСХОДНАЯ МАТРИЦА 4×4 ===")
print(matrica)
print(f"Размерность: {matrica.shape}")
print(f"Строк: {matrica.shape[0]}, Столбцов: {matrica.shape[1]}")
print()

# 2. Транспонируем матрицу (разные способы)
print("=== ТРАНСПОНИРОВАННАЯ МАТРИЦА ===")

# Способ 1: атрибут .T
transponirovannaya_1 = matrica.T
print("Способ 1 (matrica.T):")
print(transponirovannaya_1)
print(f"Размерность: {transponirovannaya_1.shape}")
print()

# Способ 2: функция np.transpose()
transponirovannaya_2 = np.transpose(matrica)
print("Способ 2 (np.transpose(matrica)):")
print(transponirovannaya_2)
print()

# Способ 3: метод .transpose()
transponirovannaya_3 = matrica.transpose()
print("Способ 3 (matrica.transpose()):")
print(transponirovannaya_3)
print()

# 3. Вычисляем след матрицы
print("=== СЛЕД МАТРИЦЫ (trace) ===")

# Способ 1: функция np.trace()
sled_1 = np.trace(matrica)
print(f"Способ 1 (np.trace): {sled_1}")

# Способ 2: метод .trace()
sled_2 = matrica.trace()
print(f"Способ 2 (matrica.trace()): {sled_2}")

# Способ 3: ручной расчет (для проверки)
diagonal = np.diag(matrica)
sled_ruchnoy = np.sum(diagonal)
print(f"Способ 3 (ручной расчет): {sled_ruchnoy}")
print()

# 4. Детальный анализ
print("=== ДЕТАЛЬНЫЙ АНАЛИЗ ===")

# Элементы на главной диагонали
print("Элементы на главной диагонали:")
print(diagonal)
print()

# Визуализация диагонали
print("Матрица с выделенной диагональю:")
for i in range(4):
    row = []
    for j in range(4):
        if i == j:
            row.append(f"[{matrica[i,j]}]")  # Выделяем диагональные элементы
        else:
            row.append(f" {matrica[i,j]} ")
    print(" ".join(row))
print()

# 5. Проверка свойств транспонирования
print("=== ПРОВЕРКА СВОЙСТВ ТРАНСПОНИРОВАНИЯ ===")

# Свойство 1: (A^T)^T = A
print("Свойство 1: (A^T)^T = A")
print(f"Исходная матрица = A")
print(f"(A^T)^T = \n{(matrica.T).T}")
print(f"Равны? {np.array_equal(matrica, (matrica.T).T)}")
print()

# Свойство 2: след A = след A^T
print("Свойство 2: trace(A) = trace(A^T)")
print(f"trace(A) = {np.trace(matrica)}")
print(f"trace(A^T) = {np.trace(matrica.T)}")
print(f"Равны? {np.trace(matrica) == np.trace(matrica.T)}")
print()

# 6. Сравнение исходной и транспонированной
print("=== СРАВНЕНИЕ ИСХОДНОЙ И ТРАНСПОНИРОВАННОЙ ===")
print("Исходная матрица:")
print(matrica)
print("\nТранспонированная матрица:")
print(matrica.T)
print(f"\nСимметричная? {np.array_equal(matrica, matrica.T)}")
```
