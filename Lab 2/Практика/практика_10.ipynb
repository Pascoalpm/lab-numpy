```python
# Упражнение 10 - Системы линейных уравнений
import numpy as np

# 1. Задаем систему уравнений:
#    2x + 3y - z = 5
#    4x - y + 2z = 6
#    -3x + 5y + 4z = -2

# Матрица коэффициентов A (3×3)
A = np.array([
    [2, 3, -1],
    [4, -1, 2],
    [-3, 5, 4]
])

# Вектор правой части B (3×1)
B = np.array([5, 6, -2])

print("=== СИСТЕМА ЛИНЕЙНЫХ УРАВНЕНИЙ ===")
print("2x + 3y - z = 5")
print("4x - y + 2z = 6")
print("-3x + 5y + 4z = -2")
print()

print("=== МАТРИЧНОЕ ПРЕДСТАВЛЕНИЕ ===")
print("Матрица коэффициентов A:")
print(A)
print(f"\nВектор правой части B:")
print(B)
print()

# 2. Решаем систему с помощью np.linalg.solve
print("=== РЕШЕНИЕ СИСТЕМЫ ===")
x = np.linalg.solve(A, B)
print(f"x = {x[0]:.2f}")
print(f"y = {x[1]:.2f}")
print(f"z = {x[2]:.2f}")
print()

# 3. Проверка решения
print("=== ПРОВЕРКА РЕШЕНИЯ ===")

# Вычисляем A @ x и сравниваем с B
proverka = A @ x
print("A × x =")
print(proverka)
print("\nДолжно быть равно B =")
print(B)
print()

print("Поэлементная проверка:")
print(f"2x + 3y - z = 2×{x[0]:.2f} + 3×{x[1]:.2f} - {x[2]:.2f} = {proverka[0]:.2f} (должно быть 5)")
print(f"4x - y + 2z = 4×{x[0]:.2f} - {x[1]:.2f} + 2×{x[2]:.2f} = {proverka[1]:.2f} (должно быть 6)")
print(f"-3x + 5y + 4z = -3×{x[0]:.2f} + 5×{x[1]:.2f} + 4×{x[2]:.2f} = {proverka[2]:.2f} (должно быть -2)")
print()

# Проверка с помощью np.allclose
print(f"Решение верно? {np.allclose(A @ x, B)}")
print()

# 4. Альтернативный метод: через обратную матрицу
print("=== АЛЬТЕРНАТИВНЫЙ МЕТОД (через обратную матрицу) ===")

# Находим обратную матрицу
A_obratnaya = np.linalg.inv(A)
print("Обратная матрица A^(-1):")
print(np.round(A_obratnaya, 3))
print()

# Решение: x = A^(-1) × B
x_alt = A_obratnaya @ B
print("Решение через обратную матрицу:")
print(f"x = {x_alt[0]:.2f}")
print(f"y = {x_alt[1]:.2f}")
print(f"z = {x_alt[2]:.2f}")
print()

# 5. Дополнительная информация
print("=== ДОПОЛНИТЕЛЬНАЯ ИНФОРМАЦИЯ ===")

# Определитель матрицы A
det_A = np.linalg.det(A)
print(f"Определитель матрицы A: det(A) = {det_A:.2f}")

# Ранг матрицы A
rank_A = np.linalg.matrix_rank(A)
print(f"Ранг матрицы A: {rank_A} (максимальный ранг = {A.shape[0]})")

# Число обусловленности
cond_A = np.linalg.cond(A)
print(f"Число обусловленности: {cond_A:.2f}")
print("(Показывает, насколько решение чувствительно к погрешностям)")
```
