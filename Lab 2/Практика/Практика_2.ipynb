```python
# Упражнение 2 - Работа с булевыми масками
import numpy as np

# 1. Генерируем массив с 20 случайными целыми числами от 1 до 100
np.random.seed(42)  # Для воспроизводимости результатов
originalniy_array = np.random.randint(1, 101, size=20)

print("=== ИСХОДНЫЙ МАССИВ ===")
print(originalniy_array)
print()

# 2. Находим элементы, делящиеся на 5
delitsya_na_5 = originalniy_array[originalniy_array % 5 == 0]
print("=== ЭЛЕМЕНТЫ, ДЕЛЯЩИЕСЯ НА 5 ===")
print(delitsya_na_5)
print(f"Количество: {len(delitsya_na_5)} элементов")
print()

# 3. Заменяем эти элементы на -1
izmenenniy_array = originalniy_array.copy()
izmenenniy_array[izmenenniy_array % 5 == 0] = -1

print("=== МАССИВ ПОСЛЕ ЗАМЕНЫ (делящиеся на 5 → -1) ===")
print(izmenenniy_array)
print()

# 4. Сравнение исходного и измененного массивов
print("=== СРАВНЕНИЕ ИСХОДНЫЙ × ИЗМЕНЕННЫЙ ===")
print("Исходный:  ", originalniy_array)
print("Измененный:", izmenenniy_array)
print()

# 5. Дополнительная статистика
print("=== СТАТИСТИКА ===")
print(f"Среднее исходного массива: {np.mean(originalniy_array):.2f}")
print(f"Среднее измененного массива: {np.mean(izmenenniy_array):.2f}")
print(f"Минимальный элемент (исходный): {np.min(originalniy_array)}")
print(f"Максимальный элемент (исходный): {np.max(originalniy_array)}")
```
