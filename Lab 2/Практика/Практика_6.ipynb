```python
# Упражнение 6 - Создание и базовые операции с матрицами
import numpy as np

# 1. Создаем две матрицы 3x3 со случайными целыми числами от 1 до 20
np.random.seed(42)  # Для воспроизводимости результатов
matrica_A = np.random.randint(1, 21, size=(3, 3))
matrica_B = np.random.randint(1, 21, size=(3, 3))

print("=== МАТРИЦА A ===")
print(matrica_A)
print(f"Размерность: {matrica_A.shape}")
print()

print("=== МАТРИЦА B ===")
print(matrica_B)
print(f"Размерность: {matrica_B.shape}")
print()

# 2. Вычисляем сумму матриц
summa_matric = matrica_A + matrica_B
print("=== СУММА МАТРИЦ (A + B) ===")
print(summa_matric)
print()

# 3. Вычисляем разность матриц
raznost_matric = matrica_A - matrica_B
print("=== РАЗНОСТЬ МАТРИЦ (A - B) ===")
print(raznost_matric)
print()

# 4. Вычисляем поэлементное произведение
poelementnoe_proizvedenie = matrica_A * matrica_B  # Поэлементное умножение
print("=== ПОЭЛЕМЕНТНОЕ ПРОИЗВЕДЕНИЕ (A * B) ===")
print(poelementnoe_proizvedenie)
print()

# 5. Дополнительный анализ
print("=== ДОПОЛНИТЕЛЬНЫЙ АНАЛИЗ ===")

# Статистика по матрице A
print("Статистика по матрице A:")
print(f"  Минимальный элемент: {np.min(matrica_A)}")
print(f"  Максимальный элемент: {np.max(matrica_A)}")
print(f"  Среднее значение: {np.mean(matrica_A):.2f}")
print(f"  Сумма всех элементов: {np.sum(matrica_A)}")

# Статистика по матрице B
print("\nСтатистика по матрице B:")
print(f"  Минимальный элемент: {np.min(matrica_B)}")
print(f"  Максимальный элемент: {np.max(matrica_B)}")
print(f"  Среднее значение: {np.mean(matrica_B):.2f}")
print(f"  Сумма всех элементов: {np.sum(matrica_B)}")

# Статистика по сумме
print("\nСтатистика по сумме матриц:")
print(f"  Минимальный элемент: {np.min(summa_matric)}")
print(f"  Максимальный элемент: {np.max(summa_matric)}")
print(f"  Среднее значение: {np.mean(summa_matric):.2f}")
print(f"  Сумма всех элементов: {np.sum(summa_matric)}")
print()

# 6. Проверка свойств
print("=== ПРОВЕРКА СВОЙСТВ ===")

# Коммутативность сложения: A + B = B + A
print(f"A + B = B + A? {np.array_equal(matrica_A + matrica_B, matrica_B + matrica_A)}")

# Коммутативность поэлементного умножения: A * B = B * A
print(f"A * B = B * A? {np.array_equal(matrica_A * matrica_B, matrica_B * matrica_A)}")
```
